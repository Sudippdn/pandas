### Replace in pandas

```DataFrame.replace(self, to_replace=None, value=None, inplace=False, limit=None, regex=False, method='pad')[source]```    
Replace values given in to_replace with value

````python
df1 = df.replace(value = 'Bharat', to_replace= "India", regex= True)
df1.loc[df['Country'] == 'India']
````

### Finding correlatin

```Syntax: DataFrame.corr(method='pearson', min_periods=1, numeric_only=False) ```

````python

df.corr(numeric_only=True).head()
````

````python
sns.heatmap(df.corr(numeric_only= True), annot = True)
# plt.tight_layout()
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
````

### Groupby and Finding the mean of all continents

```Syntax: Series.mean(axis=0, skipna=True, numeric_only=False, **kwargs)```

````python
df2 = df.groupby('Continent').mean(numeric_only  = True).sort_values('2022 Population', ascending = False)
df2
````

````python
# df[df['Continent'].str.contains('Oceania')]
````

````python
# df2.transpose()
df2.plot()
````

````python
df3 = df.groupby('Continent')[['1970 Population', '1980 Population',
                             '1990 Population', '2000 Population',
                             '2010 Population','2015 Population',
                             '2020 Population',
                             '2022 Population']].mean(numeric_only = True).sort_values("2022 Population", 
                             ascending = True)
df3
````
